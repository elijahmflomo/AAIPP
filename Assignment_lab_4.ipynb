```python
#  Task Description#1
#     • Zero-shot: Prompt AI to write a function that checks whether a given year is a leap year.
# Expected Output#1
#     • AI-generated function with no examples provided


def is_leap(year):
  """
  Checks if a given year is a leap year.

  Args:
    year: An integer representing the year.

  Returns:
    True if the year is a leap year, False otherwise.
  """
  return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

```

```python
year_to_check = 2025
if is_leap(year_to_check):
  print(f"{year_to_check} is a leap year.")
else:
  print(f"{year_to_check} is not a leap year.")
```

```python
# Task 2
# One-shot: Give one input-output example to guide AI in writing a function that converts centimeters to inches.

# Expected Output#2
# Function with correct conversion logic


def cm_to_inches(cm):
  """
  Converts centimeters to inches.

  Args:
    cm: The measurement in centimeters.

  Returns:
    The equivalent measurement in inches.
  """
  return cm * 0.393701
```

```python
# Task 3
# Few-shot: Provide a few input-output examples to guide AI in writing a function that formats full names as “Last, First”.

# Expected Output#3
# Function that correctly formats names

def format_name(full_name):
  """
  Formats a full name as "Last, First".

  Args:
    full_name: A string representing the full name (e.g., "Jane Doe").

  Returns:
    A string representing the formatted name (e.g., "Doe, Jane").
  """
  names = full_name.split()
  if len(names) >= 2:
    return f"{names[-1]}, {' '.join(names[:-1])}"
  else:
    return full_name # Return original name if not enough parts
```

```python
format_name("Flomo, Elijah M")
```

```python
# Task Description#4
#     • Compare zero-shot and few-shot prompts for writing a function that counts the number of vowels in a string.
# Expected Output#4
#     • Functional output and comparative reflection



# Zero-shot: Write a Python function named count_vowels that takes a single string as input and returns the total number of vowels (A, E, I, O, U, case-insensitive) found in the string.

def count_vowels(text):
  """
  Counts the number of vowels (case-insensitive) in a given string.

  Args:
    text: The input string.

  Returns:
    The total number of vowels in the string.
  """
  vowels = "AEIOUaeiou"
  count = 0
  for char in text:
    if char in vowels:
      count += 1
  return count

```

```python
count_vowels("This is AI Assistant Using Python")
```

```python
# Task 5
# Few-shot: Provide input-output examples to guide AI in writing a function that reads a .txt file and returns the number of lines.

# Expected Output#5
# Function that correctly counts lines in a file

def count_lines_in_file(filepath):
  """
  Reads a text file and returns the number of lines.

  Args:
    filepath: The path to the text file.

  Returns:
    The number of lines in the file.
  """
  try:
    with open(filepath, 'r') as f:
      return sum(1 for line in f)
  except FileNotFoundError:
    return f"Error: File not found at {filepath}"
  except Exception as e:
    return f"An error occurred: {e}"

```
